```python
# LING111 Project Phase 1: Building the ENG-ZH Dictionaries

# Step 1: Text Processing The Facebook AI ENG-ZH Database
# Sub-steps: removing all pairs w/ english words in the chinese place
# then, build a query list of english words to prompt from dictionaries
# also can enhance query list from NYU steinhardt thing?
# then figure out which of multiple translations is 'the best' (manual)
```

```python
# Basic text processing
enzhDict = {}

with open("en-zh.txt", "r", encoding='utf-8-sig') as in_file:
    for line in in_file: 
        line = line.strip()
        words = line.split()

        # For if Chinese equivalent is English
        if words[1] == words[0]:
            continue

        if words[0] in enzhDict:
            enzhDict[words[0]].append(words[1])
        else:
            enzhDict[words[0]] = [words[1]]

enzhDict
```

```python
import json

# Building a query list
enWords = list(enzhDict.keys())
enWords

with open("enWords.txt", "w", encoding='utf-8-sig') as out_file:
    out_file.write("\n".join(enWords))

with open("enzhDict.json", "w", encoding='utf-8-sig') as out_file:
    json.dump(enzhDict, out_file, ensure_ascii=False)
```
